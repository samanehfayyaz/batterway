```python
flows= pd.read_excel("Results_for_regression_route2.xlsx", sheet_name = "flows")
flows
```

```python
flows_shrt = flows.drop(columns=["Flow", "Type", "Unit"])
flows_shrt
```

```python
composition= pd.read_excel("Results_for_regression_route2.xlsx", sheet_name = "composition_T", 
                           dtype={'Battery': float, 'Nickel': float, 'Cobalt': float, 'Manganese': float})
print(composition.dtypes)
#composition = composition.T
#print(composition.dtypes)
#composition = composition.drop(["Flow", "Unit"])
composition
```

```python
row_index = 0
row_array = flows_shrt.iloc[row_index].to_numpy()
row_array
```

```python
x = np.linalg.solve(composition, row_array)
x
```

```python
np.allclose(np.dot(composition, x), row_array)
```

```python
factors = pd.DataFrame(columns=["Battery", "Nickel", "Cobalt", "Manganese"])

for ri in range(len(flows_shrt)):
    #print(ri)
    row_array = flows_shrt.iloc[ri].to_numpy()
    x = np.linalg.solve(composition, row_array)

    # Convert the array to a DataFrame (1 row, 4 columns)
    new_row = pd.DataFrame([x], columns=factors.columns)

    # Append the new row to the DataFrame
    factors = pd.concat([factors, new_row], ignore_index=True)

factors



```

```python
factors.to_excel("regression_results_route2.xlsx")
```

```python
# Ensure both DataFrames have the same number of rows
if len(flows) != len(factors):
    raise ValueError("Both DataFrames must have the same number of rows")

# Create the new DataFrame with the desired structure
lci_relative_battery = pd.DataFrame({
    'lci_id': 'route2',
    'direction': flows['Type'],
    'influencer': 'Battery',
    'influenced': flows['Flow'],
    'qty': factors['Battery'],
    'unit': 'kg'
})

lci_relative_battery.to_csv("relative_battery_route2.csv", index=False, sep=";")
lci_relative_battery
```
